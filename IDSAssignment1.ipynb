#### Question 2
Now it is your turn to make some plots using pandas. Let's start by transforming the data so that it is easier to work with. We then will look at some distributions of the data.

The CVDOW field isn't named helpfully and it is hard to see the meaning from the data alone. According to the website linked at the top of this notebook, CVDOW is actually indicating the day that events happened. 0->Sunday, 1->Monday ... 6->Saturday.

```python
import os
import zipfile
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Step 1: Define the URL and file name
data_url = 'http://www.ds100.org/fa20/resources/assets/datasets/lab04_data_fa20.zip'
file_name = 'lab04_data_fa20.zip'
data_dir = 'data'

# Step 2: Download the ZIP file using wget
!wget -O {file_name} {data_url}

# Step 3: Unzip the downloaded file
with zipfile.ZipFile(file_name, 'r') as zip_ref:
    zip_ref.extractall(data_dir)

# Step 4: List the extracted files to verify
print("Extracted files:")
print(os.listdir(data_dir))

# Step 5: Load a specific CSV file from the extracted contents
data_dir_path = Path(data_dir)  # creates a Path object that points to the data directory
csv_file_path = os.path.join(data_dir, 'Berkeley_PD_-_Calls_for_Service.csv')
calls = pd.read_csv(csv_file_path)

# Step 6: Display the column names to verify the presence of 'CVDOW'
print("Columns in the dataset:")
print(calls.columns)

# Step 7: Transform the CVDOW field into a more meaningful format
if 'CVDOW' in calls.columns:
    # Create a mapping for CVDOW values to day names
    day_mapping = {0: 'Sunday', 1: 'Monday', 2: 'Tuesday', 3: 'Wednesday',
                   4: 'Thursday', 5: 'Friday', 6: 'Saturday'}

    # Map the CVDOW values to their corresponding day names and create a new column 'DayOfWeek'
    calls['DayOfWeek'] = calls['CVDOW'].map(day_mapping)

    # Step 8: Analyze the distribution of calls by day of the week
    day_distribution = calls['DayOfWeek'].value_counts().reindex(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])

    # Step 9: Plot the distribution of events by day of the week
    plt.figure(figsize=(10, 6))
    day_distribution.plot(kind='bar', color='pink')
    plt.title('Number of Calls by Day of the Week')
    plt.xlabel('Day of the Week')
    plt.ylabel('Number of Calls')
    plt.xticks(rotation=45)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.show()
else:
    print("The column 'CVDOW' does not exist in the DataFrame.")
```
